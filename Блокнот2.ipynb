```python
%%R

install.packages("gapminder")
library(gapminder)
gapminder_df <- gapminder
```

Постройте график ВВП на душу населения (GDP per capita) по годам для США, Канады и Мексики. Используйте столбец gdpPercap для данных о ВВП на душу населения.

```python
%%R
summary(gapminder_df)
```

Код для выполнения **задания 1**

```python
%%R
library(dplyr)
library(ggplot2)
counts <- gapminder_df %>%
          group_by(lifeExp, continent, year)

ggplot(counts, aes(x = continent, y =lifeExp , color = continent, fill = continent)) +
  geom_bar(stat="identity")+ 
  theme_classic(base_size = 18) +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  ylab("Life expectancy") + 
  xlab("continent") +
  #ylim(c(0, 100)) + #
  ggtitle("Life Expectancy by years for continents") +
  facet_wrap(~year, ncol=6)
  # 
```

Код для выполнения **задания 2**

```python
%%R
library(dplyr)
library(ggplot2)
counts <- gapminder_df %>%
          group_by(gdpPercap, country, year)

  ggplot(counts, aes(x = year, y =gdpPercap , color = year, fill = year)) +
  geom_bar(stat="identity")+ 
  theme_classic(base_size = 18) +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  ylab("GDP per capita") + 
  xlab("Year") +
  ggtitle("GDP by years for Mexico") +
  facet_wrap(~country=="Mexico", ncol=1, scales="free")
  # facet_wrap(~country==c("Mexico","United States", "Canada"), ncol=1, scales="free")
  # 
```
