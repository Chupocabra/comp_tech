```python
%load_ext rpy2.ipython
```

```python
%%R
spisok = rnorm(10, 15, 3)
hist(spisok)
```

```python
%%R
sec_spisok = rnorm(10, 15, 3)
qqplot(spisok, sec_spisok, xlab = "test x", ylab = "test y", main = "Q-Q Plot")
```

```python
%%R
library(data.table)
animals <- fread("/content/sample_data/mammals.csv")
```

```python
%%R
plot(x=animals$body_mass_kg,  xlab = "number_of_animal", ylab = "mass", main = "Plot")
qqplot(x=animals$body_mass_kg, y=spisok, xlab = "mass", ylab = "mass", main = "qqPlot")
```

```python
%%R
qqplot(x=log(animals$body_mass_kg), y=log(spisok), xlab = "mass", ylab = "mass", main = "qqPlot")
hist(log(animals$body_mass_kg))
hist(log(spisok))
```
